Задание 4. Нарисуйте график окружности в полярных координатах. Задание 6. Напишите код, который будет рисовать график окружности в полярных координатах

```python
def plot_circle(r, x0, y0):
    x = np.linspace(-r+x0, r+x0, 501)
    plt.polar(x, np.sqrt(np.square(r)-np.square(x-x0))+y0)
    plt.polar(x, -np.sqrt(np.square(r)-np.square(x-x0))+y0)
    plt.show()
plot_circle(10,0,0)
```

Задание 5. Напишите код который будет переводить полярные координаты в декартовы

```python
def polar_to_decart(r,a):
    x = r*np.cos(a)
    y = r*np.sin(a)
    return x,y
print(polar_to_decart(2,3*np.pi))
```

Задание 8. Решите систему уравнений

```python
x = np.linspace(-2, 4, 201)
plt.plot(x, 1-(1-np.exp(x))/x)
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1,10) 
plt.grid(True)
plt.show()

from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x) + x*(1-y) - 1)

x1, y1 =  fsolve(equations, (-1, 3))
x2, y2 =  fsolve(equations, (3, 7))              
print (x1, y1)
print (x2, y2)
```

```python
x = np.linspace(-2, 4, 201)
plt.plot(x, np.abs(1-(1-np.exp(x))/x))
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1,10) 
plt.grid(True)
plt.show()

from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return (y - x**2 + 1, np.abs(np.exp(x) + x*(1-y) - 1))

x1, y1 =  fsolve(equations, (-1, 3))
x2, y2 =  fsolve(equations, (3, 7))              
print (x1, y1)
print (x2, y2)
```
