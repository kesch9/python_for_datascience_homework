координатами. (в программе)

```python
def length_vector(coord):
    sum = 0
    for i in coord:
        sum += np.square(i)
    return np.sqrt(sum)
        
```

### 3. Задание (в программе)
Напишите код на Python, реализующий построение графиков:
1. окружности,
2. эллипса,
3. гиперболы.

```python
def plot_circle(r, x0, y0):
    x = np.linspace(-r+x0, r+x0, 501)
    plt.plot(x, np.sqrt(np.square(r)-np.square(x-x0))+y0)
    plt.plot(x, -np.sqrt(np.square(r)-np.square(x-x0))+y0)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.axis('tight')
    plt.show()
plot_circle(2,1,1)
```

```python
def plot_ellipse(a, b, x0, y0):
    x = np.linspace(-a+x0, a+x0, 501)
    plt.plot(x, np.sqrt(1-np.square(x-x0)/np.square(a))*np.square(b))
    plt.plot(x, -np.sqrt(1-np.square(x-x0)/np.square(a))*np.square(b))
    plt.xlabel('x')
    plt.ylabel('y')
    plt.axis('tight')
    plt.show()
plot_ellipse(5,1,0,0)
```

```python
def plot_hiberbola(a, b):
    if a == 0:
        a = 1
    if b == 0:
        b = 1
    x = np.linspace(a, a+10, 100)
    plt.plot(x, np.sqrt(np.square(x)/np.square(a)-1)*np.square(b))
    plt.plot(x, -np.sqrt(np.square(x)/np.square(a)-1)*np.square(b))
    x = np.linspace(-b, -b-10, 100)
    plt.plot(x, np.sqrt(np.square(x)/np.square(b)-1)*np.square(b))
    plt.plot(x, -np.sqrt(np.square(x)/np.square(b)-1)*np.square(b))
    plt.xlabel('x')
    plt.ylabel('y')
    plt.axis('tight')
    plt.show()
plot_hiberbola(10,10)
```

### 5. Задание (в программе)
1) Нарисуйте трехмерный график двух параллельных плоскостей.
2) Нарисуйте трехмерный график двух любых поверхностей второго порядка.

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5,5,0.5)
Y = np.arange(-5,5,0.5)
X,Y = np.meshgrid(X,Y)
Z = 2*X + 5*Y
ax.plot_wireframe(X,Y,Z)
ax.plot_wireframe(X,Y,Z+10)
ax.scatter(0,0,0,'z',50,'red')
show()
```

```python
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5,5,0.5)
Y = np.arange(-5,5,0.5)
X,Y = np.meshgrid(X,Y)
Z = X**2+Y**2
ax.plot_wireframe(X,Y,Z)
ax.scatter(0,0,0,'z',50,'red')
show()
```
